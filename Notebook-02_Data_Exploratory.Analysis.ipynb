```python
# ==========================================
# Import Required Libraries
# ==========================================

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# Plot Settings
plt.style.use("ggplot")

pd.set_option("display.max_columns", None)

print("Libraries Imported Successfully ✅")
```

```python
# ==========================================
# Load Dataset
# ==========================================

df = pd.read_csv("heart.csv")

print("Dataset Loaded Successfully ✅")
```

```python
print("Rows :", df.shape[0])
print("Columns :", df.shape[1])
```

```python
df["HeartDisease"].value_counts()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x="HeartDisease",
    data=df
)

plt.title("Heart Disease Distribution")

plt.show()
```

## Observation

- The target classes are nearly balanced.
- Machine Learning models are unlikely to suffer from severe class imbalance.
- Accuracy can be used as one of the evaluation metrics, but Precision, Recall, and F1-score should also be considered.

```python
numeric_columns = df.select_dtypes(include=np.number).columns

numeric_columns
```

```python
numeric_columns = df.select_dtypes(include=np.number).columns

numeric_columns
```

```python
df[numeric_columns].hist(

    figsize=(18,12),

    bins=20

)

plt.tight_layout()

plt.show()
```

## Observation

- Check whether variables follow a normal distribution.
- Identify skewed features.
- Detect unusual peaks or gaps in the data.

```python
categorical_columns = df.select_dtypes(include="object").columns

categorical_columns
```

```python
for column in categorical_columns:

    plt.figure(figsize=(6,4))

    sns.countplot(
        x=column,
        data=df
    )

    plt.title(column)

    plt.xticks(rotation=20)

    plt.show()
```

## Observation

- Identify dominant categories.
- Check whether any category has very few samples.
- This helps in deciding encoding techniques later.

```python
for column in numeric_columns:

    plt.figure(figsize=(6,3))

    sns.boxplot(
        x=df[column]
    )

    plt.title(column)

    plt.show()
```

## Observation

- Boxplots help detect outliers.
- Outlier treatment will be decided in the preprocessing notebook.
- Not every outlier should be removed, especially in medical datasets.


```python
corr = df.corr(numeric_only=True)

corr
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```

## Observation

- Highly correlated features may introduce multicollinearity.
- Feature selection may be required based on correlation and model performance.

```python
plt.figure(figsize=(7,4))

sns.boxplot(
    x="HeartDisease",
    y="Age",
    data=df
)

plt.show()
```

```python
plt.figure(figsize=(7,4))

sns.boxplot(
    x="HeartDisease",
    y="Oldpeak",
    data=df
)

plt.show()
```

```python
plt.figure(figsize=(7,4))

sns.boxplot(
    x="HeartDisease",
    y="MaxHR",
    data=df
)

plt.show()
```

```python
sns.pairplot(

    df,

    hue="HeartDisease",

    diag_kind="hist"

)

plt.show()
```
