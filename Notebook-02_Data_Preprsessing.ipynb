# Notebook 03 : Data Preprocessing

## Objective

This notebook prepares the dataset for machine learning.

### Tasks

- Missing Value Analysis
- Duplicate Check
- Feature Encoding
- Feature Scaling
- Train-Test Split

> Note:
> No model will be trained in this notebook.

```python
# =====================================
# Import Required Libraries
# =====================================

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

print("Libraries Imported Successfully ✅")
```

```python
# =====================================
# Load Dataset
# =====================================

df = pd.read_csv("heart.csv")

df.head()
```

```python
df.info()
```

```python
missing_values = df.isnull().sum()

print("Missing Values")

missing_values
```

```python
duplicate_rows = df.duplicated().sum()

print("Duplicate Rows :", duplicate_rows)
```

```python
df = df.drop_duplicates()

print("New Shape :", df.shape)
```

```python
X = df.drop("HeartDisease", axis=1)

y = df["HeartDisease"]

print("Features Shape :", X.shape)

print("Target Shape :", y.shape)
```

```python
categorical_columns = X.select_dtypes(include="object").columns

categorical_columns
```

```python
X = pd.get_dummies(

    X,

    columns=categorical_columns,

    drop_first=True

)

X.head()
```

```python
print("Encoded Feature Shape :", X.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(

    X,

    y,

    test_size=0.20,

    random_state=42,

    stratify=y

)

print("Training Samples :", X_train.shape)

print("Testing Samples :", X_test.shape)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)

print("Scaling Completed Successfully ✅")
```

```python
X_train_scaled = pd.DataFrame(

    X_train_scaled,

    columns=X_train.columns

)

X_test_scaled = pd.DataFrame(

    X_test_scaled,

    columns=X_test.columns

)
```

```python
X_train_scaled.head()
```

```python
print("="*60)

print("Data Preprocessing Completed Successfully")

print("="*60)

print("Original Dataset :", df.shape)

print("Training Set :", X_train_scaled.shape)

print("Testing Set :", X_test_scaled.shape)

print("="*60)
```
